## 0. Read in the data

```python
def read_csv(filename):
    """ Reads read-only, binary, comma-sep csv files, input name with ' ' """
    import csv
    with open(filename,'rU') as csvfile:
        filereader = csv.reader(csvfile, delimiter=',')
        for row in filereader:
            print ', '.join(row)    
```
